```python
import os
import pathlib
import pandas as pd
import numpy as np
import sys
```

```python
df=pd.read_csv("C:/Users/srg_minitower/ibeems_data/df_hdc_2024.csv")
```

```python
cum_cols=df.columns[df.columns.str.contains("cum_")]
```

```python
cum_cols
```

```python
for ccols in cum_cols:
    df[ccols]=df[ccols].diff().bfill()
```

```python
import matplotlib.pyplot as plt
```

```python
df.index=pd.to_datetime(df['timestamp_local'])
```

```python
fig,ax=plt.subplots(2,1,figsize=(15,5))
ax[0].plot(df.index,df['V_ahu_hcw_ahu9']/3600,label='V_ahu_hcw_ahu9',linewidth=1.5)
ax[0].plot(df.index,df['cum_ahu_hcw_ahu9'].rolling(20,min_periods=1,center=True).mean().to_numpy()/1000/30,label='cum_ahu_hcw_ahu9',linewidth=0.5)
ax0=ax[0].twinx()
ax0.plot(df.index,df['T_ahu_hcws_ahu9'],label="T_ahu_hcws_ahu9",color='tab:green')
ax0.plot(df.index,df['T_ahu_hcwr_ahu9'],label="T_ahu_hcwr_ahu9",color='tab:red')
ax[1].plot(df.index,df['V_ahu_hcw_ahu11']/3600,label='V_ahu_hcw_ahu11',linewidth=1.5)
ax[1].plot(df.index,df['cum_ahu_hcw_ahu11'].rolling(20,min_periods=1,center=True).mean().to_numpy()/1000/30,label='cum_ahu_hcw_ahu11',linewidth=0.5)
ax1=ax[1].twinx()
ax1.plot(df.index,df['T_ahu_hcws_ahu11'],label="T_ahu_hcws_ahu11",color='tab:green')
ax1.plot(df.index,df['T_ahu_hcwr_ahu11'],label="T_ahu_hcwr_ahu11",color='tab:red')
ax[0].legend()
ax0.legend()
ax[1].legend()
ax[0].set_ylabel("Flowrate m3/s")
ax[1].set_ylabel("Flowrate m3/s")
```
